```python

import re
from collections import Counter

# Входной список ссылок
# links = '''https://www.instagram.com/findgoodsongs?igsh=OHczNm1rczd6d3pv
# https://www.instagram.com/themusicfeeds?igsh=bmNqYmtycDJ1YnU3
# https://www.instagram.com/muzicpeaces?igsh=MXdzN3I4cXF2dDRsdw==
# https://www.instagram.com/unexpected.songs?igsh=b3kyM3M2Y2h6NDRx
# https://www.instagram.com/itsjmaine?igsh=enh3enVtbGxweHJ5
# https://www.instagram.com/rapmusic?igsh=MXVpZm1pcXRrbWx1Nw==
# https://www.instagram.com/reel/DFR3cy9NgqZ/?igsh=cGZsbWhuejlpNjlp
# https://www.tiktok.com/@dailydoseraps?_t=ZM-8tWcSYNAQa5&_r=1
# https://vm.tiktok.com/ZMkV6kSS9/
# https://vm.tiktok.com/ZMkV6FUNd/
# '''

# Instagram usernames extraction
try:
    # Подсчет количества входящих ссылок Instagram
    instagram_links = [line for line in links.strip().split('\n') if 'instagram.com' in line]
    total_instagram_links = len(instagram_links)

    # Извлечение usernames Instagram, исключая видео-ссылки
    instagram_usernames = []
    instagram_video_links = []
    for line in instagram_links:
        if "instagram.com/reel/" in line or "instagram.com/p/" in line:
            instagram_video_links.append(line)
        else:
            match = re.search(r'(?:https?://(?:www\.)?instagram\.com/)([\w\.]+)', line)
            if match:
                instagram_usernames.append(match.group(1))

    # Подсчет повторений Instagram
    instagram_username_counts = Counter(instagram_usernames)

    # Вывод всех извлеченных usernames Instagram
    #print("\nВсе извлеченные usernames Instagram:")
    # for user in instagram_usernames:
    #     print(user)

    # Вывод повторяющихся с количеством Instagram
    print("\nПовторяющиеся usernames Instagram и количество повторений:")
    found_instagram_repeat = False
    for user, count in instagram_username_counts.items():
        if count > 1:
            print(f"{user}: {count}")
            found_instagram_repeat = True
    if not found_instagram_repeat:
        print("Нет повторяющихся usernames Instagram")

    # Уникальные usernames Instagram (без дубликатов)
    unique_instagram_usernames = list(instagram_username_counts.keys())
    print("\nВсе уникальные usernames Instagram (без дубликатов):")
    print("\n".join(unique_instagram_usernames))

    # Вывод метрик Instagram
    print(f"\nКоличество входящих ссылок Instagram: {total_instagram_links}")
    print(f"Количество уникальных usernames Instagram: {len(unique_instagram_usernames)}")

    # Вывод ссылок на видео Instagram
    print("\nСсылки на видео Instagram:")
    if instagram_video_links:
        for video in instagram_video_links:
            print(video)
    else:
        print("Нет ссылок на видео Instagram")

except Exception as e:
    print(f"Ошибка Instagram: {e}")

# TikTok usernames extraction с фильтрацией видео
try:
    # Списки для записей
    tiktok_usernames = []
    tiktok_video_links = []
    
    # Подсчет количества входящих ссылок TikTok
    tiktok_links = [line for line in links.strip().split('\n') if 'tiktok.com' in line]
    total_tiktok_links = len(tiktok_links)
    
    # Обход каждой TikTok-ссылки
    for line in tiktok_links:
        # Если ссылка содержит "vm.tiktok.com", то это видео
        if "vm.tiktok.com" in line:
            tiktok_video_links.append(line)
            continue

        # Попытка извлечения username из ссылки вида "https://www.tiktok.com/@username..."
        match = re.search(r'(?:https?://(?:www\.)?tiktok\.com/@)([\w\.]+)', line)
        if match:
            username = match.group(1)
            # Если username равен "reel" или "p", то считаем, что это видео
            if username in ('reel', 'p'):
                tiktok_video_links.append(line)
            else:
                tiktok_usernames.append(username)
    
    # Подсчет повторений TikTok usernames
    tiktok_username_counts = Counter(tiktok_usernames)
    
    # Вывод всех извлеченных usernames TikTok
    #print("\nВсе извлеченные usernames TikTok (без video-ссылок):")
    # for user in tiktok_usernames:
    #     print(user)
    
    # Вывод повторяющихся usernames TikTok, если такие есть
    print("\nПовторяющиеся usernames TikTok и количество повторений:")
    found_tiktok_repeat = False
    for user, count in tiktok_username_counts.items():
        if count > 1:
            print(f"{user}: {count}")
            found_tiktok_repeat = True
    if not found_tiktok_repeat:
        print("Нет повторяющихся usernames TikTok")
    
    # Уникальные usernames TikTok (без дубликатов)
    unique_tiktok_usernames = list(tiktok_username_counts.keys())
    print("\nВсе уникальные usernames TikTok (без дубликатов):")
    print("\n".join(unique_tiktok_usernames))
    
    # Вывод метрик TikTok для usernames
    print(f"\nКоличество входящих ссылок TikTok: {total_tiktok_links}")
    print(f"Количество уникальных usernames TikTok: {len(unique_tiktok_usernames)}")
    
    # Вывод ссылок на видео (TikTok video-ссылки)
    print("\nСсылки на видео TikTok:")
    if tiktok_video_links:
        for video in tiktok_video_links:
            print(video)
    else:
        print("Нет ссылок на видео TikTok")
        
except Exception as e:
    print(f"Ошибка TikTok: {e}")
```
